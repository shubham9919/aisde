```python
from os import walk
import json
filenames = next(walk('./'), (None, None, []))[2]  # [] if no file
filenames = [file for file in filenames if file.endswith('domain_api.json')]
filenames
```

```python
final_res = []
for filename in filenames:
    file = open(filename, encoding="utf-8")
    data = json.load(file)
    res = dict()
    company_name = data['Results'][0]['Meta']['CompanyName']
    res['company'] = company_name if company_name else filename[:-16]
    res['technologies'] = [child['Name'] for child in data['Results'][0]['Result']['Paths'][0]['Technologies']]
    final_res.append(res)
    
final_res
```

```python
with open('BuiltWithData3.json', 'w') as fp:
    json.dump(final_res, fp, indent=4)
```
